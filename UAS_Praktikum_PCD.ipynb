```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Menentukan path file gambar pada Google Colab
image_path = '/content/roko.jpg'

# Membaca gambar
image = cv2.imread(image_path)

# Mendefinisikan titik-titik koordinat sumber
pts1 = np.float32([[243, 481], [591, 486], [210, 1114], [681, 1101]])

# Mendefinisikan titik-titik koordinat tujuan
pts2 = np.float32([[0, 0], [400, 0], [0, 600], [400, 600]])

# Mendapatkan matriks transformasi perspektif
matrix = cv2.getPerspectiveTransform(pts1, pts2)

# Menerapkan transformasi perspektif
warped_image = cv2.warpPerspective(image, matrix, (400, 600))

# Menampilkan gambar asli dan gambar yang telah di-transformasi
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Menampilkan gambar asli
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')

# Menampilkan gambar yang telah di-transformasi
ax2.imshow(cv2.cvtColor(warped_image, cv2.COLOR_BGR2RGB))
ax2.set_title('Transformed Image')

plt.show()
```
